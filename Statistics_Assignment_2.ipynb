```python
#4.	Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

sample_data = [45, 48, 50, 49, 46, 51, 47, 50, 52, 49]
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
sigma = np.std(sample_data)


population_mean = 50 #( Known population mean)

# Calculate the Z-score
z_score = (sample_mean - population_mean) / (sigma / np.sqrt(sample_size))

# Find the p-value for the two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Display results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sigma}")
print(f"Z-Score: {z_score}")
print(f"P-Value: {p_value}")

# Interpret the result
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")


# Plotting the Z-distribution and decision region
# Critical value for two-tailed test (alpha=0.05)
critical_value = stats.norm.ppf(1 - alpha / 2)

# Range for plotting the Z-distribution
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

# Plot the Z-distribution
plt.plot(x, y, label='Z-dist.')

# Shade the rejection regions
plt.fill_between(x, 0, y, where=(x <= -critical_value), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, 0, y, where=(x >= critical_value), color='red', alpha=0.3, label='Rejection Region')

# Mark the critical values and Z-statistic
plt.axvline(x=-critical_value, color='black', linestyle='--', label=f'-Critical Value')
plt.axvline(x=critical_value, color='black', linestyle='--', label=f'+Critical Value')


# Labels and Title
plt.title('Two-Tailed Z-Test')
plt.xlabel('Z value')
plt.ylabel('Prob. Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#5.	Create a Python function that calculates and visualizes Type 1 and  Type 2 errors during hypothesis testing.


def visualize_errors(mu0, mu1, sigma, alpha, beta, show_plot=True):
    """
    Calculates and visualizes Type 1 and Type 2 errors in hypothesis testing.

    Args:
        mu0: Mean under the null hypothesis.
        mu1: Mean under the alternative hypothesis.
        sigma: Standard deviation.
        alpha: Significance level (Type I error rate).
        beta: Type II error rate.

    """

    # Calculate critical value (z-score) for Type I error
    z_alpha = stats.norm.ppf(1 - alpha)

    # Calculate critical value (z-score) for Type II error
    z_beta = stats.norm.ppf(1 - beta)

    # Calculate decision boundary
    decision_boundary = mu0 + z_alpha * sigma

    # Calculate power (1 - beta)
    power = 1 - beta

    # Create x-axis values for plotting
    x = np.linspace(mu0 - 4 * sigma, mu1 + 4 * sigma, 1000)

    # Calculate probability density functions
    pdf_null = stats.norm.pdf(x, mu0, sigma)
    pdf_alt = stats.norm.pdf(x, mu1, sigma)

    # Plot the distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf_null, label='Null Hypothesis')
    plt.plot(x, pdf_alt, label='Alternative Hypothesis')

    # Shade the rejection region (Type I error)
    plt.fill_between(x, 0, pdf_null, where=(x > decision_boundary), color='red', alpha=0.25, label='Type I Error')

    # Shade the acceptance region under the alternative (Type II error)
    plt.fill_between(x, 0, pdf_alt, where=(x < decision_boundary), color='green', alpha=0.25, label='Type II Error')

    # Add vertical lines for decision boundary and means
    plt.axvline(mu0, color='blue', linestyle='--', label='mu0')
    plt.axvline(mu1, color='green', linestyle='--', label='mu1')
    plt.axvline(decision_boundary, color='red', linestyle='--', label='Decision Boundary')

    # Add labels and title
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.title('Type I and Type II Error Visualization')
    plt.legend()
    plt.show()

    # Print summary
    print(f'Significance Level (alpha): {alpha}')
    print(f'Type II Error Rate (beta): {beta}')
    print(f'Power (1 - beta): {power}')

# Example usage
mu0 = 0  # Mean under the null hypothesis
mu1 = 2  # Mean under the alternative hypothesis
sigma = 1  # Standard deviation
alpha = 0.05  # Significance level
beta = 0.2  # Type II error rate

visualize_errors(mu0, mu1, sigma, alpha, beta)

```

```python
#9.	Write a Python function to calculate the confidence interval for a sample mean and explain its significance.


def confidence_interval(data, confidence_level=0.95):

    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation

    # Calculate the critical value (t-value for small samples, z-value for large samples)
    if n < 30:  # Use t-distribution for small samples
        t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)
    else:  # Use standard normal distribution for large samples
        t_critical = stats.norm.ppf((1 + confidence_level) / 2)

    # Calculate the margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

data = [10, 12, 15, 18, 20]
confidence_interval = confidence_interval(data)
print(f"Confidence Interval: {confidence_interval}")
```

```python
#10.	Write a Python program to calculate the margin of error for a given confidence level using sample data.


def calculate_moe(data, confidence_level=0.95):
    n = len(data)
    sample_std = np.std(data)

    if n < 30:  # Use t-distribution for small samples
        critical_val = stats.t.ppf((1 + confidence_level) / 2, df=n-1)
    else:  # Use standard normal distribution for large samples
        critical_val = stats.norm.ppf((1 + confidence_level) / 2)

    # Calculate the margin of error
    margin_of_error = critical_val * (sample_std / np.sqrt(n))

    return margin_of_error

data = [10, 12, 15, 18, 20]
margin_of_error = calculate_moe(data)
print(f"Margin of Error: {margin_of_error}")
```

```python
#16.	Implement an F-test using Python to compare the variances of two random samples.

sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std Dev=10
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std Dev=15

# Calculate sample variances
var1 = np.var(sample1)
var2 = np.var(sample2)

# Compute the F-statistic
F_stat = var1 / var2 if var1 > var2 else var2 / var1  # Ensure F > 1

# Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Compute p-value (two-tailed test)
p_value = 2 * (1 - stats.f.cdf(F_stat, df1, df2))

# Print results
print(f"Sample 1 Variance: {var1:.4f}")
print(f"Sample 2 Variance: {var2:.4f}")
print(f"F-statistic: {F_stat:.4f}")
print(f"p-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

```python
#19.	Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

def check_anova_assumptions(data, groups):
  results = {}

  # Check normality within each group
  normality_results = []
  for group_data in data:
    w, p_value = stats.shapiro(group_data)
    normality_results.append(p_value)
  results['normality'] = normality_results

  # Check for equal variances (homogeneity of variances)
  _, p_value_levene = stats.levene(*data)
  results['equal_variance'] = p_value_levene

  # Independence: Usually assessed based on study design
  results['independence'] = "Assumed based on study design. Please verify based on your specific experiment."

  return results

# Example usage
data = [[10, 12, 15, 18, 20], [8, 10, 11, 14, 16], [12, 14, 16, 18, 20]]
groups = ['Group 1', 'Group 2', 'Group 3']

assumption_results = check_anova_assumptions(data, groups)

print("Normality p-values:", assumption_results['normality'])
print("Equal variance p-value:", assumption_results['equal_variance'])
print("Independence:", assumption_results['independence'])
```

```python
#20.	Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import statsmodels.formula.api as sm
from statsmodels.stats.anova import anova_lm

def perform_two_way_anova(data, formula, alpha=0.05):

    model = sm.ols(formula, data=data).fit()
    anova_table = anova_lm(model, typ=2)

    print("Two-Way ANOVA Table:")
    print(anova_table)

    # Correct way to access the interaction p-value:
    interaction_term = 'C(Factor1):C(Factor2)'
    if interaction_term in anova_table.index:
        interaction_p = anova_table['PR(>F)'][interaction_term]
        if interaction_p < alpha:
            print(f"\nSignificant interaction between Factor1 and Factor2 (p = {interaction_p:.3f}).")
        else:
            print(f"\nNo significant interaction between Factor1 and Factor2 (p = {interaction_p:.3f}).")
    else:
        print("\nInteraction term not found in ANOVA table (check your formula).")


    # Interaction Plot
    sns.pointplot(x='Factor1', y='DV', hue='Factor2', data=data, capsize=.2, join=True)
    plt.title('Interaction Plot')
    plt.xlabel('Factor1')
    plt.ylabel('Dependent Variable (DV)')
    plt.show()

    #Main effects plots
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Factor1', y='DV', data=data)
    plt.title('Main Effect of Factor1')
    plt.xlabel('Factor1')
    plt.ylabel('Dependent Variable (DV)')
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Factor2', y='DV', data=data)
    plt.title('Main Effect of Factor2')
    plt.xlabel('Factor2')
    plt.ylabel('Dependent Variable (DV)')
    plt.show()


    return anova_table


# Example Usage :
data = pd.DataFrame({
    'DV': np.random.rand(100),
    'Factor1': ['A', 'B'] * 50,
    'Factor2': ['X', 'Y'] * 25 * 2
})


formula = 'DV ~ C(Factor1) * C(Factor2)'
anova_table = perform_two_way_anova(data, formula)
```

```python
#21.	Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.


def visualize_f_distribution(df1, df2, alpha=0.05):

    x = np.linspace(0, stats.f.ppf(0.999, df1, df2), 500)  # x values for the plot
    y = stats.f.pdf(x, df1, df2)  # F-distribution PDF

    plt.plot(x, y, label=f"F({df1}, {df2}) Distribution")

    # Shade the critical region (right tail)
    critical_value = stats.f.ppf(1 - alpha, df1, df2)  # Critical F-value
    x_critical = np.linspace(critical_value, x.max(), 200)
    plt.fill_between(x_critical, stats.f.pdf(x_critical, df1, df2), color='red', alpha=0.5, label=f"Critical Region (α = {alpha})")

    plt.title("F-Distribution Visualization")
    plt.xlabel("F-statistic")
    plt.ylabel("Probability Density")
    plt.axvline(critical_value, color='red', linestyle='--', label=f"Critical Value = {critical_value:.2f}")  # Vertical line at critical value
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Degrees of Freedom (Numerator): {df1}")
    print(f"Degrees of Freedom (Denominator): {df2}")
    print(f"Critical F-value (for α = {alpha}): {critical_value:.2f}")


# Example usage:
df1 = 5
df2 = 20
visualize_f_distribution(df1, df2)


df1 = 2
df2 = 10
visualize_f_distribution(df1, df2, alpha=0.01)
```

```python
#25.	Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

# Function to perform Z-test for two proportions
def ztest_two_proportions(x1, n1, x2, n2, alpha=0.05):
    # Sample proportions
    p1 = x1 / n1
    p2 = x2 / n2

    # Pooled proportion
    pooled_p = (x1 + x2) / (n1 + n2)

    # Standard error
    se = np.sqrt(pooled_p * (1 - pooled_p) * (1/n1 + 1/n2))

    # Z-test statistic
    z_stat = (p1 - p2) / se

    # Two-tailed p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    # Critical value for two-tailed test
    critical_value = stats.norm.ppf(1 - alpha / 2)

    # Output results
    print(f"Sample Proportion 1: {p1:.4f}")
    print(f"Sample Proportion 2: {p2:.4f}")
    print(f"Pooled Proportion: {pooled_p:.4f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical Value (α = {alpha}): ±{critical_value:.4f}")

    # Interpret the results
    if abs(z_stat) > critical_value:
        print("Reject the null hypothesis: The proportions are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The proportions are not significantly different.")

# Example data
x1 = 45  # Number of successes in group 1
n1 = 100  # Sample size of group 1
x2 = 55  # Number of successes in group 2
n2 = 150  # Sample size of group 2

# Perform the Z-test
ztest_two_proportions(x1, n1, x2, n2)
```

```python
#26.	Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

# Example datasets
data1 = np.array([24, 25, 28, 23, 22, 30, 29])
data2 = np.array([31, 32, 35, 30, 28, 36, 34])

# Calculate variances
var1 = np.var(data1, ddof=1)  # ddof=1 for sample variance
var2 = np.var(data2, ddof=1)

# Ensure var1 is the larger variance
if var1 < var2:
    var1, var2 = var2, var1
    data1, data2 = data2, data1

# Calculate F-statistic
F = var1 / var2

# Degrees of freedom
df1 = len(data1) - 1
df2 = len(data2) - 1

# Calculate p-value
p_value = 2 * min(stats.f.cdf(F, df1, df2), 1 - stats.f.cdf(F, df1, df2))

# Output results
print(f"Variance of Dataset 1: {var1:.2f}")
print(f"Variance of Dataset 2: {var2:.2f}")
print(f"F-statistic: {F:.2f}")
print(f"Degrees of freedom (df1, df2): ({df1}, {df2})")
print(f"P-value: {p_value:.4f}")

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: Variances are not significantly different.")

# Visualization
x = np.linspace(0, 5, 1000)
f_dist = stats.f.pdf(x, df1, df2)
plt.plot(x, f_dist, label=f"F-distribution (df1={df1}, df2={df2})")
plt.axvline(F, color='red', linestyle='--', label=f"F-statistic = {F:.2f}")
plt.fill_between(x, f_dist, where=(x >= F), color='red', alpha=0.3, label="p-value region")
plt.title("F-Distribution and F-Statistic")
plt.xlabel("F-value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
