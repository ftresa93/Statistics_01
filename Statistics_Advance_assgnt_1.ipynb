Q13.	What is the Central Limit Theorem (CLT), and why is it critical in statistics?

A13. The Central Limit Theorem  states that if you have a population with mean mu and std dev sigma and sufficiently large number of random samples from the population with replacement then the sampling distribution of the sample mean becomes approximately normal as n increases.

Sampling distribution mean will be mu and std dev or standard error will sigma/sqrt(n), n is the number of members in the sample.

Why is the CLT Critical in Statistics?
1. CLT Enables Inferential Statistics: The CLT allows us to make inferences about population parameters using sample data by leveraging the normal distribution, whose properties we know.

2. Justifies Hypothesis Testing: Many statistical tests (e.g., t-tests, z-tests) assume that sample means follow a normal distribution, which is supported by the CLT for large enough samples.

3. Facilitates Confidence Intervals: The normality of the sampling distribution makes it possible to construct confidence intervals for population parameters.


4. Works with Real-World Data:Most real-world populations are not normally distributed, yet the CLT ensures that sample means can still be analyzed using normal-based methods.The CLT reduces the complexity of analyzing distributions with unknown or non-normal shapes by ensuring a normal approximation for large sample sizes

Q17.	What are point estimates and interval estimates in statistics?

A17. A point estimate is a single value that is used as an estimate of an unknown population parameter. For example, If you take a sample of 100 people and calculate the sample mean height to be 170 cm, then 170 cm is a point estimate of the population mean height.

An interval estimate provides a range of values, rather than a single value, within which the population parameter is likely to lie. For example, based on the sample, the population mean height is estimated to lie between 168 cm and 172 cm with 95% confidence. Here the range is an interval estimate, and the "95% confidence" indicates that if the sampling process were repeated many times, 95% of the resulting intervals would contain the true population mean.

```python
#18.	Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.


def calculate_z_score(data_point, mean, std_dev):
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_standard_normal(z_score):
    probability = stats.norm.cdf(z_score)  # Cumulative probability
    return probability

# Example data
data_point = 75
sample_data = [60, 65, 70, 72, 74, 76, 78, 100, 85, 90]  # Sample dataset

# Calculate mean and standard deviation
mean = np.mean(sample_data)
std_dev = np.std(sample_data, ddof=1)  # Using sample standard deviation

# Compute Z-score
z_score = calculate_z_score(data_point, mean, std_dev)

# Compare to standard normal distribution
probability = compare_to_standard_normal(z_score)

print(f"Data Point: {data_point}")
print(f"Mean: {mean:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"Probability from Standard Normal Distribution: {probability:.4f}")
```

```python
#19.	Implement hypothesis testing using Z-statistics for a sample dataset.

def z_test(sample_data, population_mean, population_std, alpha=0.05, alternative='two-tailed'):

    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Compute Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Compute p-value
    if alternative == 'two-tailed':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
        critical_value = stats.norm.ppf(1 - alpha / 2)
    elif alternative == 'left-tailed':
        p_value = stats.norm.cdf(z_score)
        critical_value = stats.norm.ppf(alpha)
    elif alternative == 'right-tailed':
        p_value = 1 - stats.norm.cdf(z_score)
        critical_value = stats.norm.ppf(1 - alpha)
    else:
        raise ValueError("Alternative hypothesis must be 'two-tailed', 'left-tailed', or 'right-tailed'")

    # Decision rule
    if (alternative == 'two-tailed' and abs(z_score) > critical_value) or (alternative == 'left-tailed' and z_score < critical_value) or (alternative == 'right-tailed' and z_score > critical_value):
        decision = "Reject H0"
    else:
        decision = "Fail to Reject H0"

    return z_score, p_value, decision

# Example Usage
sample_data = [50, 52, 48, 49, 51, 50, 52, 53, 48, 49]
population_mean = 50  # Hypothesized mean
population_std = 2    # Known population standard deviation
alpha = 0.05         # Significance level
alternative = 'two-tailed'  # Type of test

z_score, p_value, decision = z_test(sample_data, population_mean, population_std, alpha, alternative)

print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Decision: {decision}")
```

```python
#20.	Create a confidence interval for a dataset using Python and interpret the result.


def compute_confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    margin_of_error = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)

    return mean - margin_of_error, mean + margin_of_error

# Example dataset
data = [12, 15, 14, 10, 13, 16, 18, 14, 15, 17, 19, 14, 16, 12, 15]

# Compute 95% confidence interval
lower_bound, upper_bound = compute_confidence_interval(data)

# Print results
print(f"Mean of dataset: {np.mean(data):.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Interpretation
print("\nInterpretation:")
print(f"We are 95% confident that the true mean of the population from which this sample was drawn lies between {lower_bound:.2f} and {upper_bound:.2f}.")
```

```python
#25.	Write a Python program to compare the standard deviations of two datasets and visualize the difference.


def generate_data(mean1=50, std1=10, mean2=50, std2=20, size=1000):
    """Generates two datasets with specified mean and standard deviation."""
    data1 = np.random.normal(mean1, std1, size)
    data2 = np.random.normal(mean2, std2, size)
    return data1, data2

def compare_standard_deviations(data1, data2):
    """Calculates and prints the standard deviations of two datasets."""
    std1 = np.std(data1, ddof=1)
    std2 = np.std(data2, ddof=1)
    print(f"Standard Deviation of Dataset 1: {std1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std2:.2f}")
    return std1, std2

def plot_distributions(data1, data2):
    """Plots histograms and boxplots to visualize dataset distributions."""
    plt.figure(figsize=(12, 6))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data1, bins=30, kde=True, color='blue', label='Dataset 1', alpha=0.6)
    sns.histplot(data2, bins=30, kde=True, color='red', label='Dataset 2', alpha=0.6)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.title("Histogram of Datasets")
    plt.legend()
    plt.show()

# Generate datasets
data1, data2 = generate_data()

# Compare standard deviations
compare_standard_deviations(data1, data2)

# Visualize distributions
plot_distributions(data1, data2)
```

```python
#26.	Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

data = np.random.normal(loc=50, scale=15, size=1000)

# Calculate the range
data_range = np.max(data) - np.min(data)

# Calculate the interquartile range (IQR)
Q1 = np.percentile(data, 25)  # First quartile (25th percentile)
Q3 = np.percentile(data, 75)  # Third quartile (75th percentile)
iqr = Q3 - Q1


print("data range is:",data_range)
print("the IQR is: ",iqr)
```

```python
#28.	Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.


def calculate_skewness_kurtosis(data):
    skewness = stats.skew(data)
    kurt = stats.kurtosis(data, fisher=True)
    return skewness, kurt

data = np.random.normal(loc=50, scale=15, size=1000)

# Calculate skewness and kurtosis
skewness, kurt = calculate_skewness_kurtosis(data)
print("Skewness",skewness)
print("Kurtosis",kurt)
```
